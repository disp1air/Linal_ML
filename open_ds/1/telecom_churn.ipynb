Для категориальных(тип object) и булевых(тип bool) признаков можно воспользоваться методом **value_counts**, который показывает
как распределяется признак, т.е. кол-во объектов с указанным значением

```python
df['Churn'].value_counts()
```

Т.е. у 2850 пользователей из 3333 в свойстве 'Churn' указано False

**Сортировка**

DataFrame можно отсортировать по значению какого-нибудь из признаков.
В данном случае по 'Total day charge'. (ascending=False для сортировки по убыванию): 

```python
df.sort_values(by='Total day charge', ascending=False).head()
```
