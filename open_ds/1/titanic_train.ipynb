```python
def first_name(str):
    name_list = str.split(' ')
    if name_list[-2] == 'Mr.':
        return name_list[-1]
    elif name_list[-2] == 'Master.':
        return name_list[-1]
    else:
        return name_list[-2]
```

```python
data['First_male_name'] = data[data['Sex'] == 'male']['Name'].apply(first_name)
```

```python
data['First_male_name'].value_counts().head(10) # Десять самых популярных мужских имен
```

```python
# Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания?
for i, sub_df in data.groupby('Pclass'):
    print(sub_df[sub_df['Sex'] == 'male']['Age'].mean())
```

```python
for i, sub_df in data.groupby('Pclass'):
    print(sub_df[sub_df['Sex'] == 'female']['Age'].mean())
```
